```python
import pandas as pd
```

# 沪深300——基本股价数据

```python
hs300 =  pd.read_csv('../../data/final/hs300(2013-2022).csv').set_index('trade_date').sort_index()
hs300.index = pd.to_datetime(hs300.index, format='%Y%m%d')
hs300
```

# 宏观数据


```python
# CPI
CPI = pd.read_csv("../../data/final/CPI.csv",encoding='gbk')
CPI.columns = ['date','CPI']
CPI.set_index('date',inplace=True)
CPI.index = pd.to_datetime(CPI.index)
CPI = CPI['20130101':'20221231'].resample('M').mean()
CPI.index = CPI.index.strftime('%Y%m')
CPI
```

```python
# PPI
PPI = pd.read_csv("../../data/final/PPI.csv",encoding='gbk')
PPI.columns = ['date','PPI']
PPI.set_index('date',inplace=True)
PPI.index = pd.to_datetime(PPI.index)
PPI = PPI['20130101':'20221231'].resample('M').mean()
PPI.index = PPI.index.strftime('%Y%m')
PPI
```

```python
# PMI
PMI = pd.read_csv("../../data/final/PMI.csv",encoding='gbk')
PMI.columns = ['date','PMI']
PMI.set_index('date',inplace=True)
PMI.index = pd.to_datetime(PMI.index)
PMI = PMI['20130101':'20221231'].resample('M').mean()
PMI.index = PMI.index.strftime('%Y%m')
PMI
```

```python
# 社会消费品零售总额
retail = pd.read_csv("../../data/final/社零.csv",encoding='gbk')
retail.columns = ['date','retail']
retail.set_index('date',inplace=True)
retail.index = pd.to_datetime(retail.index)
retail = retail['20130101':'20221231'].resample('M').mean()
retail.index = retail.index.strftime('%Y%m')
retail = retail.reindex(PMI.index)
retail
```

```python
# 市场利率
rate = pd.read_csv("../../data/final/利率.csv",encoding='gbk')
rate.columns = ['date','rate']
rate.set_index('date',inplace=True)
rate.index = pd.to_datetime(rate.index)
rate = rate['20130101':'20221231'].resample('M').mean()
rate.index = rate.index.strftime('%Y%m')
rate = rate.reindex(PMI.index)
rate
```

```python
# 社会融资
Social_Financing = pd.read_csv("../../data/final/社会融资.csv",encoding='gbk')
Social_Financing.columns = ['date','SF']
Social_Financing.set_index('date',inplace=True)
Social_Financing.index = pd.to_datetime(Social_Financing.index)
Social_Financing = Social_Financing['20130101':'20221231'].resample('M').mean()
Social_Financing.index = Social_Financing.index.strftime('%Y%m')
Social_Financing = Social_Financing.reindex(PMI.index)
Social_Financing
```

```python
# 国债收益率2年
Bond_Yield2 = pd.read_csv("../../data/final/国债收益率(2年).csv",encoding='gbk')
Bond_Yield2.columns = ['date','BY2']
Bond_Yield2.set_index('date',inplace=True)
Bond_Yield2.index = pd.to_datetime(Bond_Yield2.index)
Bond_Yield2 = Bond_Yield2['20130101':'20221231'].resample('M').mean()
Bond_Yield2.index = Bond_Yield2.index.strftime('%Y%m')
Bond_Yield2 = Bond_Yield2.reindex(PMI.index)
Bond_Yield2
```

```python
# 国债收益率10年
Bond_Yield10 = pd.read_csv("../../data/final/国债收益率(10年).csv",encoding='gbk')
Bond_Yield10.columns = ['date','BY10']
Bond_Yield10.set_index('date',inplace=True)
Bond_Yield10.index = pd.to_datetime(Bond_Yield10.index)
Bond_Yield10 = Bond_Yield10['20130101':'20221231'].resample('M').mean()
Bond_Yield10.index = Bond_Yield10.index.strftime('%Y%m')
Bond_Yield10 = Bond_Yield10.reindex(PMI.index)
Bond_Yield10
```

```python
# 美元指数
Dollar_Index = pd.read_csv("../../data/final/美元指数.csv",encoding='gbk')
Dollar_Index.columns = ['date','DI']
Dollar_Index.set_index('date',inplace=True)
Dollar_Index.index = pd.to_datetime(Dollar_Index.index)
Dollar_Index = Dollar_Index['20130101':'20221231'].resample('M').mean()
Dollar_Index.index = Dollar_Index.index.strftime('%Y%m')
Dollar_Index = Dollar_Index.reindex(PMI.index)
Dollar_Index
```

```python
# 汇率
Exchange_Rate = pd.read_csv("../../data/final/汇率USD_CNY.csv",encoding='gbk')
Exchange_Rate.columns = ['date','ER']
Exchange_Rate.set_index('date',inplace=True)
Exchange_Rate.index = pd.to_datetime(Exchange_Rate.index)
Exchange_Rate = Exchange_Rate['20130101':'20221231'].resample('M').mean()
Exchange_Rate.index = Exchange_Rate.index.strftime('%Y%m')
Exchange_Rate = Exchange_Rate.reindex(PMI.index)
Exchange_Rate
```

```python
# 大宗商品价格指数CCPI
CCPI = pd.read_csv("../../data/final/CCPI.csv",encoding='gbk')
CCPI.columns = ['date','CCPI']
CCPI.set_index('date',inplace=True)
CCPI.index = pd.to_datetime(CCPI.index)
CCPI = CCPI['20130101':'20221231'].resample('M').mean()
CCPI.index = CCPI.index.strftime('%Y%m')
CCPI = CCPI.reindex(PMI.index)
CCPI
```

```python
# 通过hs300，提取平均月收益率
hs_index = hs300.copy()
hs_index['return'] = hs_index['close'].pct_change()
hs_index['return_t+1'] = hs_index['return'].shift(-1)
hs_index['return_t+1'] = hs_index['return_t+1'].fillna(0)
hs_index['return_t+1'] = hs_index['return_t+1'].apply(lambda x:round(x,4))
hs_index['return_t+1(%)'] = hs_index['return_t+1']*100
hs_index = hs_index['20130101':'20221231']
hs_index = hs_index.resample('M').mean()
hs_index.index = hs_index.index.strftime('%Y%m')
hs_index
```

```python
# 合并数据
Macro_data = pd.concat([CPI,PPI,PMI,retail,rate,Social_Financing,Bond_Yield2,Bond_Yield10,Dollar_Index,Exchange_Rate,CCPI, hs_index["return_t+1(%)"]],axis=1)
Macro_data = Macro_data.fillna(Macro_data.mean())
Macro_data
```
